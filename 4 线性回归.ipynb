### 训练模型

```python
# 导入数据
from sklearn import datasets 
boston = datasets.load_boston()
# 导入LinearRegression类创建一个对象
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# 将训练集传给lr拟合
lr.fit(boston.data,boston.target)
```

### 评估模型

```python
predictions = lr.predict(boston.data)
```

```python
# 残差图
import numpy as np
residual = np.histogram(boston.target-predictions,bins=50)
from pyecharts import Bar
bar = Bar()
bar.add('',residual[1].tolist()[:-1],residual[0].tolist())
bar
```
