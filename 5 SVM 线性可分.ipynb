* 可视化超平面 1

```python
x,y = datasets.make_blobs(n_features=2,centers=2)
from sklearn.svm import LinearSVC
svm = LinearSVC()
svm.fit(x,y)
```

```python
from itertools import product 
from collections import namedtuple
import numpy as np
```

```python
Point = namedtuple('Point',['x', 'y', 'outcome'])
decision_boundary=[]
xmin,xmax=np.percentile(x[:,0],[0,100])
ymin,ymax=np.percentile(x[:,1],[0,100])
```

```python
for xpt,ypt in product(np.linspace(xmin-2.5,xmax+2.5,20),np.linspace(ymin-2.5,ymax+2.5,20)):
    p = Point(xpt,ypt,svm.predict([xpt,	ypt]))
    decision_boundary.append(p)
```

```python
from pyecharts import Scatter
color_list = ['#006400','#8B4513']
scatter = Scatter()
class_0 = np.array([[point.x,point.y] for point in decision_boundary if point.outcome[0]==0])
class_1 = np.array([[point.x,point.y] for point in decision_boundary if point.outcome[0]==1])
scatter.add('',class_0[:,0],class_0[:,1],symbol_size=2)
scatter.add('',class_1[:,0],class_1[:,1],symbol_size=2)
scatter.add('',x[y==0][:,0],x[y==0][:,1])
scatter.add('',x[y==1][:,0],x[y==1][:,1],is_xaxis_show=0,is_yaxis_show=0)
scatter
```

* 可视化超平面 2

```python
x,y = datasets.make_classification(n_features=2,
                                   n_classes=2,
                                   n_informative=2,
                                   n_redundant=0)
svm.fit(x,y)
xmin,xmax = np.percentile(x[:,0],[0,100])
ymin,ymax = np.percentile(x[:,1],[0,100])
test_points = np.array([[xx,yy]for  xx,yy in product(np.linspace(xmin,xmax),np.linspace(ymin,ymax))])
test_preds = svm.predict(test_points)

from pyecharts import Scatter
color_lst = ['#FFFFFF','#000000',' #006400','#008B8B']
scatter = Scatter()
scatter.add('',test_points[test_preds==0][:,0],test_points[test_preds==0][:,1],symbol_size=2)#,label_color=['red']
scatter.add('',test_points[test_preds==1][:,0],test_points[test_preds==1][:,1],symbol_size=2)
scatter.add('',x[y==0][:,0],x[y==0][:,1],symbol_size=15)
scatter.add('',x[y==1][:,0],x[y==1][:,1],is_xaxis_show=0,is_yaxis_show=0,symbol_size=15)
scatter
```
