```python
from sklearn.svm import SVC
from sklearn import datasets
import numpy as np
import pyecharts
import warnings
from itertools import product
warnings.filterwarnings('ignore')
```

```python
x,y =datasets.make_classification(n_features=2,
                                   n_classes=2,
                                   n_informative=2,
                                   n_redundant=0)
radial_svm = SVC(kernel='rbf')
radial_svm.fit(x,y)
np.percentile(x[:,0],[0,100])

xmin,xmax = np.percentile(x[:,0],[0,100])
ymin,ymax = np.percentile(x[:,1],[0,100])
test_points = np.array([[xx,yy]for xx,yy in product(np.linspace(xmin,xmax,num=60),np.linspace(ymin,ymax,num=60))])
test_preds = radial_svm.predict(test_points)
scatter = pyecharts.Scatter()
scatter.add('',test_points[test_preds==0][:,0],test_points[test_preds==0][:,1],symbol_size=3)
scatter.add('',test_points[test_preds==1][:,0],test_points[test_preds==1][:,1],symbol_size=3)
scatter.add('',x[y==0][:,0],x[y==0][:,1],symbol_size=15)
scatter.add('',x[y==1][:,0],x[y==1][:,1],symbol_size=15,is_xaxis_show=False,is_yaxis_show=False)
scatter
```
